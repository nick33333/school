# Assignment 3: Problem 8
<br>
For problem 8 of assignment 3, I needed to generate the k-mer composition <br>
given some number k and a sequence of nucleotides.<br>

```python
'''
Problem 8: Generate the k-mer Composition of a String
'''
class Solution:
    '''
    Class carries the solution to Problem 8. Given an integer k
    and a string, Solution generates the k-mers that can be produced
    '''
    def __init__(self, text:'str', k:'str'):
        '''
        Initializes Solution object with input text, input integer k, and a
        list of the k-mer compositions from the text.
        '''
        self.text = text
        self.k = k
        self.compList = self.kmerComp()
        
    def kmerComp(self)->'list':
        '''
        Using input text, create a list of all possible compositions
        of length k and return it.
        '''
        frame = len(self.text)-self.k+1
        kmerList = [] # to hold list of k-mer compositions of size k
        for pos in range(frame):
            kmerList.append(self.text[pos:pos+self.k])
        kmerList.sort()
        return kmerList
```

# Main function 
Main function is written here. This function handles argument parsing, input, and output.

```python
def main(infile, outfile, inCL=None):
    '''
    The main function reads in some integer k which represents the size
    of the k-mer compositions to generate. The main function then reads
    in a sequence of nucleotides in the form of a string which is to be
    converted into a k-mer composition.
    '''
    k=0
    inputSeq = ""
    
    with open(infile,'r') as myfile:
        for line in myfile:
            myLine = line.rstrip()
            k = int(myLine)
            print(f"k:{k}")
            inputSeq += next(myfile).rstrip()    
                
    solution = Solution(inputSeq, k)
    solutionList = solution.compList
    
    with open(outfile, "w") as myfile:
        for kmer in solutionList:
            myfile.write(kmer+'\n')
            print(kmer)
```

```python
if __name__ == "__main__":
    main(infile='data/rosalind_ba3a.txt',outfile='output/output-problem8.txt', inCL=[])
```
